```python
import os
import ast
import json
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    auc
)
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout,
    BatchNormalization
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# Paths
# ============================================================

input_file = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time\Food_Delivery_Time_Prediction.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time"

os.makedirs(output_dir, exist_ok=True)

print("Output directory:")
print(output_dir)

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(input_file)

print("\nDataset Loaded Successfully")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

# ============================================================
# Helper Functions
# ============================================================

def parse_location(location):
    try:
        loc = ast.literal_eval(str(location))
        return pd.Series([float(loc[0]), float(loc[1])])
    except:
        return pd.Series([np.nan, np.nan])


def haversine(lat1, lon1, lat2, lon2):
    R = 6371

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        np.sin(dlat / 2) ** 2
        + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    )

    c = 2 * np.arcsin(np.sqrt(a))

    return R * c


def create_rush_hour_feature(time_value):
    try:
        hour = pd.to_datetime(str(time_value)).hour

        if (8 <= hour <= 10) or (18 <= hour <= 21):
            return 1
        else:
            return 0

    except:
        return 0


def tabular_to_image(data):
    n_samples, n_features = data.shape

    image_size = int(np.ceil(np.sqrt(n_features)))
    total_pixels = image_size * image_size

    padded_data = np.zeros((n_samples, total_pixels))
    padded_data[:, :n_features] = data

    image_data = padded_data.reshape(n_samples, image_size, image_size, 1)

    return image_data, image_size


def build_cnn_model(
    image_size,
    filters=32,
    kernel_size=3,
    learning_rate=0.001,
    dense_units=64,
    dropout_rate=0.3
):
    model = Sequential()

    model.add(
        Conv2D(
            filters=filters,
            kernel_size=(kernel_size, kernel_size),
            activation="relu",
            padding="same",
            input_shape=(image_size, image_size, 1)
        )
    )

    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(
        Conv2D(
            filters=filters * 2,
            kernel_size=(kernel_size, kernel_size),
            activation="relu",
            padding="same"
        )
    )

    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(Flatten())
    model.add(Dense(dense_units, activation="relu"))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation="sigmoid"))

    optimizer = Adam(learning_rate=learning_rate)

    model.compile(
        optimizer=optimizer,
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model


# ============================================================
# Phase 1: Data Preprocessing
# ============================================================

print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

if "Order_ID" in df.columns:
    df.drop(columns=["Order_ID"], inplace=True)

# ============================================================
# Feature Engineering
# ============================================================

if "Customer_Location" in df.columns:
    df[["Customer_Latitude", "Customer_Longitude"]] = df["Customer_Location"].apply(parse_location)

if "Restaurant_Location" in df.columns:
    df[["Restaurant_Latitude", "Restaurant_Longitude"]] = df["Restaurant_Location"].apply(parse_location)

if all(col in df.columns for col in [
    "Customer_Latitude",
    "Customer_Longitude",
    "Restaurant_Latitude",
    "Restaurant_Longitude"
]):
    df["Haversine_Distance"] = haversine(
        df["Customer_Latitude"],
        df["Customer_Longitude"],
        df["Restaurant_Latitude"],
        df["Restaurant_Longitude"]
    )

if "Order_Time" in df.columns:
    df["Rush_Hour"] = df["Order_Time"].apply(create_rush_hour_feature)
    df["Order_Hour"] = pd.to_datetime(df["Order_Time"], errors="coerce").dt.hour
    df["Order_Hour"] = df["Order_Hour"].fillna(df["Order_Hour"].median())

# ============================================================
# Target Variable Creation
# ============================================================

median_delivery_time = df["Delivery_Time"].median()

df["Delivery_Status"] = np.where(
    df["Delivery_Time"] <= median_delivery_time,
    "Fast",
    "Delayed"
)

print("\nMedian Delivery Time:", median_delivery_time)

target_encoder = LabelEncoder()
df["Delivery_Status_Encoded"] = target_encoder.fit_transform(df["Delivery_Status"])

print("\nTarget Classes:")
for label, value in zip(
    target_encoder.classes_,
    target_encoder.transform(target_encoder.classes_)
):
    print(label, "=", value)

# ============================================================
# Remove Unnecessary Raw Columns
# ============================================================

drop_columns = [
    "Customer_Location",
    "Restaurant_Location",
    "Order_Time",
    "Delivery_Status"
]

existing_drop_columns = [col for col in drop_columns if col in df.columns]

df_model = df.drop(columns=existing_drop_columns)

X = df_model.drop(columns=["Delivery_Status_Encoded"])
y = df_model["Delivery_Status_Encoded"]

# ============================================================
# Identify Column Types
# ============================================================

numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()

print("\nNumerical Columns:")
print(numeric_cols)

print("\nCategorical Columns:")
print(categorical_cols)

# ============================================================
# Preprocessing Pipeline
# ============================================================

numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

X_processed = preprocessor.fit_transform(X)

if hasattr(X_processed, "toarray"):
    X_processed = X_processed.toarray()

print("\nProcessed Feature Shape:", X_processed.shape)

# ============================================================
# Convert Tabular Data to CNN Image Format
# ============================================================

X_images, image_size = tabular_to_image(X_processed)

print("\nCNN Input Image Shape:", X_images.shape)
print("Image Size:", image_size, "x", image_size)

# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_images,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(
    X_processed,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# ============================================================
# Phase 2: CNN Model Training
# ============================================================

cnn_model = build_cnn_model(image_size=image_size)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

history = cnn_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

cnn_model.save(os.path.join(output_dir, "cnn_food_delivery_model.h5"))

# ============================================================
# CNN Evaluation
# ============================================================

y_pred_prob = cnn_model.predict(X_test).ravel()
y_pred = (y_pred_prob >= 0.5).astype(int)

cnn_accuracy = accuracy_score(y_test, y_pred)
cnn_precision = precision_score(y_test, y_pred)
cnn_recall = recall_score(y_test, y_pred)
cnn_f1 = f1_score(y_test, y_pred)

print("\nCNN Model Performance")
print("Accuracy:", cnn_accuracy)
print("Precision:", cnn_precision)
print("Recall:", cnn_recall)
print("F1 Score:", cnn_f1)

report_text = classification_report(
    y_test,
    y_pred,
    target_names=target_encoder.classes_
)

print("\nClassification Report:")
print(report_text)

with open(os.path.join(output_dir, "cnn_classification_report.txt"), "w") as f:
    f.write(report_text)

# ============================================================
# Confusion Matrix
# ============================================================

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
plt.imshow(cm)
plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.colorbar()

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xticks([0, 1], target_encoder.classes_)
plt.yticks([0, 1], target_encoder.classes_)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_confusion_matrix.png"))
plt.show()

# ============================================================
# ROC Curve
# ============================================================

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, label=f"CNN AUC = {roc_auc:.4f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.title("CNN ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_roc_curve.png"))
plt.show()

# ============================================================
# Training Accuracy Graph
# ============================================================

plt.figure(figsize=(7, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("CNN Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_accuracy.png"))
plt.show()

# ============================================================
# Training Loss Graph
# ============================================================

plt.figure(figsize=(7, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CNN Training and Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_loss.png"))
plt.show()

# ============================================================
# Logistic Regression Baseline Model
# ============================================================

logistic_model = LogisticRegression(max_iter=1000)

logistic_model.fit(X_train_lr, y_train_lr)

lr_pred = logistic_model.predict(X_test_lr)
lr_prob = logistic_model.predict_proba(X_test_lr)[:, 1]

lr_accuracy = accuracy_score(y_test_lr, lr_pred)
lr_precision = precision_score(y_test_lr, lr_pred)
lr_recall = recall_score(y_test_lr, lr_pred)
lr_f1 = f1_score(y_test_lr, lr_pred)

print("\nLogistic Regression Performance")
print("Accuracy:", lr_accuracy)
print("Precision:", lr_precision)
print("Recall:", lr_recall)
print("F1 Score:", lr_f1)

# ============================================================
# Model Comparison
# ============================================================

model_comparison = pd.DataFrame(
    {
        "Model": ["CNN", "Logistic Regression"],
        "Accuracy": [cnn_accuracy, lr_accuracy],
        "Precision": [cnn_precision, lr_precision],
        "Recall": [cnn_recall, lr_recall],
        "F1_Score": [cnn_f1, lr_f1]
    }
)

model_comparison.to_csv(
    os.path.join(output_dir, "model_comparison_results.csv"),
    index=False
)

plt.figure(figsize=(8, 5))

x = np.arange(len(model_comparison["Model"]))
width = 0.2

plt.bar(x - width, model_comparison["Accuracy"], width, label="Accuracy")
plt.bar(x, model_comparison["Precision"], width, label="Precision")
plt.bar(x + width, model_comparison["Recall"], width, label="Recall")
plt.bar(x + 2 * width, model_comparison["F1_Score"], width, label="F1 Score")

plt.xticks(x, model_comparison["Model"])
plt.title("CNN vs Logistic Regression Performance")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "model_comparison_graph.png"))
plt.show()

# ============================================================
# Phase 3: 5-Fold Cross Validation for CNN
# ============================================================

kfold = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

cv_results = []
fold = 1

for train_index, val_index in kfold.split(X_images, y):

    print("\nTraining Fold", fold)

    X_train_fold = X_images[train_index]
    X_val_fold = X_images[val_index]

    y_train_fold = y.iloc[train_index]
    y_val_fold = y.iloc[val_index]

    fold_model = build_cnn_model(image_size=image_size)

    fold_early_stop = EarlyStopping(
        monitor="val_loss",
        patience=8,
        restore_best_weights=True
    )

    fold_model.fit(
        X_train_fold,
        y_train_fold,
        validation_data=(X_val_fold, y_val_fold),
        epochs=50,
        batch_size=16,
        callbacks=[fold_early_stop],
        verbose=0
    )

    val_prob = fold_model.predict(X_val_fold).ravel()
    val_pred = (val_prob >= 0.5).astype(int)

    fold_accuracy = accuracy_score(y_val_fold, val_pred)
    fold_precision = precision_score(y_val_fold, val_pred)
    fold_recall = recall_score(y_val_fold, val_pred)
    fold_f1 = f1_score(y_val_fold, val_pred)

    cv_results.append(
        {
            "Fold": fold,
            "Accuracy": fold_accuracy,
            "Precision": fold_precision,
            "Recall": fold_recall,
            "F1_Score": fold_f1
        }
    )

    print("Fold Accuracy:", fold_accuracy)

    fold += 1

cv_results_df = pd.DataFrame(cv_results)

cv_results_df.to_csv(
    os.path.join(output_dir, "cnn_cross_validation_results.csv"),
    index=False
)

print("\nCNN Cross Validation Results")
print(cv_results_df)

print("\nAverage CV Accuracy:", cv_results_df["Accuracy"].mean())
print("Average CV F1 Score:", cv_results_df["F1_Score"].mean())

plt.figure(figsize=(7, 5))
plt.plot(
    cv_results_df["Fold"],
    cv_results_df["Accuracy"],
    marker="o",
    label="Accuracy"
)
plt.plot(
    cv_results_df["Fold"],
    cv_results_df["F1_Score"],
    marker="o",
    label="F1 Score"
)

plt.title("CNN 5-Fold Cross Validation Performance")
plt.xlabel("Fold")
plt.ylabel("Score")
plt.xticks(cv_results_df["Fold"])
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_cross_validation_graph.png"))
plt.show()

# ============================================================
# CNN Hyperparameter Tuning
# ============================================================

param_grid = [
    {
        "filters": 16,
        "kernel_size": 2,
        "learning_rate": 0.001,
        "dense_units": 32,
        "dropout_rate": 0.2
    },
    {
        "filters": 32,
        "kernel_size": 3,
        "learning_rate": 0.001,
        "dense_units": 64,
        "dropout_rate": 0.3
    },
    {
        "filters": 64,
        "kernel_size": 3,
        "learning_rate": 0.0005,
        "dense_units": 128,
        "dropout_rate": 0.4
    }
]

tuning_results = []

for params in param_grid:

    print("\nTesting CNN Parameters:")
    print(params)

    temp_model = build_cnn_model(
        image_size=image_size,
        filters=params["filters"],
        kernel_size=params["kernel_size"],
        learning_rate=params["learning_rate"],
        dense_units=params["dense_units"],
        dropout_rate=params["dropout_rate"]
    )

    temp_early_stop = EarlyStopping(
        monitor="val_loss",
        patience=8,
        restore_best_weights=True
    )

    temp_model.fit(
        X_train,
        y_train,
        validation_split=0.2,
        epochs=50,
        batch_size=16,
        callbacks=[temp_early_stop],
        verbose=0
    )

    temp_prob = temp_model.predict(X_test).ravel()
    temp_pred = (temp_prob >= 0.5).astype(int)

    temp_accuracy = accuracy_score(y_test, temp_pred)
    temp_precision = precision_score(y_test, temp_pred)
    temp_recall = recall_score(y_test, temp_pred)
    temp_f1 = f1_score(y_test, temp_pred)

    result = params.copy()

    result.update(
        {
            "Accuracy": temp_accuracy,
            "Precision": temp_precision,
            "Recall": temp_recall,
            "F1_Score": temp_f1
        }
    )

    tuning_results.append(result)

tuning_results_df = pd.DataFrame(tuning_results)

tuning_results_df.to_csv(
    os.path.join(output_dir, "cnn_hyperparameter_tuning_results.csv"),
    index=False
)

best_result = tuning_results_df.sort_values(
    by="F1_Score",
    ascending=False
).iloc[0]

print("\nCNN Hyperparameter Tuning Results")
print(tuning_results_df)

print("\nBest CNN Parameters")
print(best_result)

# ============================================================
# Save Final Predictions
# ============================================================

final_predictions = pd.DataFrame(
    {
        "Actual_Label": target_encoder.inverse_transform(y_test),
        "Predicted_Label": target_encoder.inverse_transform(y_pred),
        "Prediction_Probability": y_pred_prob
    }
)

final_predictions.to_csv(
    os.path.join(output_dir, "cnn_predictions.csv"),
    index=False
)

# ============================================================
# Final Report
# ============================================================

final_report = f"""
Food Delivery Time Prediction using CNN and Model Validation

Author:
Sagnik Patra

Objective:
The goal of this project is to predict whether a food delivery will be Fast or Delayed
based on customer location, restaurant location, weather conditions, traffic conditions,
delivery person experience, order priority, ratings, vehicle type, distance, and other
delivery-related features.

Dataset:
Food_Delivery_Time_Prediction.csv

Dataset Shape:
{df.shape}

Target Variable:
The target variable was created from Delivery_Time.

Fast:
Delivery_Time <= Median Delivery_Time

Delayed:
Delivery_Time > Median Delivery_Time

Median Delivery Time:
{median_delivery_time}

Preprocessing:
1. Missing numerical values were handled using median imputation.
2. Missing categorical values were handled using most frequent imputation.
3. Categorical variables were encoded using One-Hot Encoding.
4. Numerical variables were normalized using StandardScaler.
5. Location coordinates were extracted from customer and restaurant locations.

Feature Engineering:
1. Customer latitude and longitude were extracted.
2. Restaurant latitude and longitude were extracted.
3. Haversine distance was calculated.
4. Order hour was extracted.
5. Rush hour feature was created.
6. Delivery status was created as Fast or Delayed.

CNN Method:
The dataset was tabular, so the processed feature vector was converted into an
image-like square matrix. This allowed the CNN model to process the features using
Conv2D layers.

CNN Architecture:
1. Conv2D Layer
2. Batch Normalization
3. MaxPooling2D
4. Conv2D Layer
5. Batch Normalization
6. MaxPooling2D
7. Flatten Layer
8. Dense Layer
9. Dropout Layer
10. Sigmoid Output Layer

CNN Performance:
Accuracy: {cnn_accuracy}
Precision: {cnn_precision}
Recall: {cnn_recall}
F1 Score: {cnn_f1}
ROC AUC: {roc_auc}

Logistic Regression Performance:
Accuracy: {lr_accuracy}
Precision: {lr_precision}
Recall: {lr_recall}
F1 Score: {lr_f1}

Cross Validation:
Number of Folds: 5

Average CV Accuracy:
{cv_results_df["Accuracy"].mean()}

Average CV Precision:
{cv_results_df["Precision"].mean()}

Average CV Recall:
{cv_results_df["Recall"].mean()}

Average CV F1 Score:
{cv_results_df["F1_Score"].mean()}

Best CNN Hyperparameters:
{best_result.to_dict()}

Generated Assignment Files:
1. cnn_food_delivery_model.h5
2. cnn_classification_report.txt
3. model_comparison_results.csv
4. cnn_cross_validation_results.csv
5. cnn_hyperparameter_tuning_results.csv
6. cnn_predictions.csv
7. cnn_confusion_matrix.png
8. cnn_roc_curve.png
9. cnn_training_accuracy.png
10. cnn_training_loss.png
11. cnn_cross_validation_graph.png
12. model_comparison_graph.png
13. final_report.txt

Excluded From Submission:
1. Dataset CSV file
2. Raw data dumps
3. Environment folders
4. Cache folders
5. Build folders
6. Notebook outputs should be cleared before submission

Conclusion:
The CNN model was successfully developed and evaluated for food delivery classification.
The system predicts whether a delivery is Fast or Delayed using delivery, location,
weather, traffic, and time-based features. The model was validated using train-test
evaluation, confusion matrix, ROC curve, Logistic Regression comparison, 5-fold
cross-validation, and hyperparameter tuning.
"""

with open(os.path.join(output_dir, "final_report.txt"), "w") as f:
    f.write(final_report)

# ============================================================
# Submission Checklist
# ============================================================

required_submission_files = [
    "Food_Delivery_Time_CNN.ipynb",
    "README.md",
    "final_report.txt",
    "cnn_food_delivery_model.h5",
    "cnn_classification_report.txt",
    "model_comparison_results.csv",
    "cnn_cross_validation_results.csv",
    "cnn_hyperparameter_tuning_results.csv",
    "cnn_predictions.csv",
    "cnn_confusion_matrix.png",
    "cnn_roc_curve.png",
    "cnn_training_accuracy.png",
    "cnn_training_loss.png",
    "cnn_cross_validation_graph.png",
    "model_comparison_graph.png"
]

excluded_files = [
    "Food_Delivery_Time_Prediction.csv",
    "processed_food_delivery_data.csv",
    "dataset_preview.csv",
    "venv/",
    "env/",
    ".venv/",
    "node_modules/",
    "dist/",
    "build/",
    ".next/",
    ".cache/",
    "__pycache__/",
    ".ipynb_checkpoints/"
]

print("\n====================================================")
print("PROJECT COMPLETED SUCCESSFULLY")
print("====================================================")

print("\nGenerated files are saved in:")
print(output_dir)

print("\nSubmit only these files:")
for file in required_submission_files:
    file_path = os.path.join(output_dir, file)

    if os.path.exists(file_path):
        print("Available:", file)
    else:
        print("Create/Check:", file)

print("\nDo NOT submit these files/folders:")
for item in excluded_files:
    print("Exclude:", item)

print("\nImportant:")
print("Before submission, clear large notebook outputs manually from Jupyter Notebook.")
print("Do not include dataset, raw processed dumps, virtual environments, caches, or build folders.")
```
