```python
import os
import ast
import json
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_curve, auc
)
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# Paths
# ============================================================

input_file = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time\Food_Delivery_Time_Prediction.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time"

os.makedirs(output_dir, exist_ok=True)

print("Output directory:", output_dir)

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(input_file)

print("\nDataset Loaded Successfully")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

# Save basic dataset info
df.head().to_csv(os.path.join(output_dir, "dataset_preview.csv"), index=False)

# ============================================================
# Helper Functions
# ============================================================

def parse_location(location):
    """
    Convert location string like '(17.030479, 79.743077)' into latitude and longitude.
    """
    try:
        loc = ast.literal_eval(str(location))
        return pd.Series([float(loc[0]), float(loc[1])])
    except:
        return pd.Series([np.nan, np.nan])


def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate geographical distance between two latitude-longitude points.
    """
    R = 6371
    lat1, lon1, lat2, lon2 = map(
        np.radians, [lat1, lon1, lat2, lon2]
    )

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        np.sin(dlat / 2) ** 2
        + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    )

    c = 2 * np.arcsin(np.sqrt(a))
    return R * c


def create_rush_hour_feature(time_value):
    """
    Rush hour:
    Morning: 8 AM - 10 AM
    Evening: 6 PM - 9 PM
    """
    try:
        hour = pd.to_datetime(str(time_value)).hour
        if (8 <= hour <= 10) or (18 <= hour <= 21):
            return 1
        return 0
    except:
        return 0


def delivery_label(time):
    """
    Create target label:
    Fast = 0
    Delayed = 1
    Based on median delivery time.
    """
    if time <= median_delivery_time:
        return "Fast"
    else:
        return "Delayed"


# ============================================================
# Phase 1: Data Preprocessing and Feature Engineering
# ============================================================

print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

# Drop Order_ID because it is not useful for prediction
if "Order_ID" in df.columns:
    df.drop(columns=["Order_ID"], inplace=True)

# Extract coordinates
df[["Customer_Latitude", "Customer_Longitude"]] = df["Customer_Location"].apply(parse_location)
df[["Restaurant_Latitude", "Restaurant_Longitude"]] = df["Restaurant_Location"].apply(parse_location)

# Calculate Haversine distance
df["Haversine_Distance"] = haversine(
    df["Customer_Latitude"],
    df["Customer_Longitude"],
    df["Restaurant_Latitude"],
    df["Restaurant_Longitude"]
)

# Rush hour feature
df["Rush_Hour"] = df["Order_Time"].apply(create_rush_hour_feature)

# Time features
df["Order_Hour"] = pd.to_datetime(df["Order_Time"], errors="coerce").dt.hour
df["Order_Hour"] = df["Order_Hour"].fillna(df["Order_Hour"].median())

# Create target variable
median_delivery_time = df["Delivery_Time"].median()
df["Delivery_Status"] = df["Delivery_Time"].apply(delivery_label)

print("\nMedian Delivery Time:", median_delivery_time)

# Encode target
target_encoder = LabelEncoder()
df["Delivery_Status_Encoded"] = target_encoder.fit_transform(df["Delivery_Status"])

print("\nTarget Classes:")
for cls, enc in zip(target_encoder.classes_, target_encoder.transform(target_encoder.classes_)):
    print(cls, "=", enc)

# Drop original location and time text columns
drop_columns = [
    "Customer_Location",
    "Restaurant_Location",
    "Order_Time",
    "Delivery_Status"
]

df_model = df.drop(columns=drop_columns)

# Separate X and y
X = df_model.drop(columns=["Delivery_Status_Encoded"])
y = df_model["Delivery_Status_Encoded"]

# Identify columns
numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()

print("\nNumerical Columns:")
print(numeric_cols)

print("\nCategorical Columns:")
print(categorical_cols)

# Preprocessing
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

X_processed = preprocessor.fit_transform(X)

if hasattr(X_processed, "toarray"):
    X_processed = X_processed.toarray()

print("\nProcessed Feature Shape:", X_processed.shape)

# Save processed dataset
processed_df = pd.DataFrame(X_processed)
processed_df["Delivery_Status_Encoded"] = y.values
processed_df.to_csv(os.path.join(output_dir, "processed_food_delivery_data.csv"), index=False)

# ============================================================
# Convert Tabular Data into Image-like CNN Input
# ============================================================

def tabular_to_image(data):
    """
    Convert tabular data into square image format for CNN.
    Example:
    If features are 25, convert to 5x5 image.
    If not perfect square, pad with zeros.
    """
    n_samples, n_features = data.shape
    image_size = int(np.ceil(np.sqrt(n_features)))
    total_pixels = image_size * image_size

    padded_data = np.zeros((n_samples, total_pixels))
    padded_data[:, :n_features] = data

    image_data = padded_data.reshape(n_samples, image_size, image_size, 1)

    return image_data, image_size


X_images, image_size = tabular_to_image(X_processed)

print("\nCNN Image Shape:", X_images.shape)

# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_images,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("\nTraining Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)

# ============================================================
# Phase 2: CNN Model
# ============================================================

def build_cnn_model(
    image_size,
    filters=32,
    kernel_size=3,
    learning_rate=0.001,
    dense_units=64,
    dropout_rate=0.3
):
    model = Sequential()

    model.add(Conv2D(
        filters=filters,
        kernel_size=(kernel_size, kernel_size),
        activation="relu",
        padding="same",
        input_shape=(image_size, image_size, 1)
    ))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(Conv2D(
        filters=filters * 2,
        kernel_size=(kernel_size, kernel_size),
        activation="relu",
        padding="same"
    ))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(Flatten())

    model.add(Dense(dense_units, activation="relu"))
    model.add(Dropout(dropout_rate))

    model.add(Dense(1, activation="sigmoid"))

    optimizer = Adam(learning_rate=learning_rate)

    model.compile(
        optimizer=optimizer,
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model


cnn_model = build_cnn_model(image_size=image_size)

cnn_model.summary()

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

history = cnn_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

# Save CNN model
cnn_model.save(os.path.join(output_dir, "cnn_food_delivery_model.h5"))

# ============================================================
# CNN Evaluation
# ============================================================

y_pred_prob = cnn_model.predict(X_test).ravel()
y_pred = (y_pred_prob >= 0.5).astype(int)

cnn_accuracy = accuracy_score(y_test, y_pred)
cnn_precision = precision_score(y_test, y_pred)
cnn_recall = recall_score(y_test, y_pred)
cnn_f1 = f1_score(y_test, y_pred)

print("\nCNN Evaluation Results")
print("Accuracy:", cnn_accuracy)
print("Precision:", cnn_precision)
print("Recall:", cnn_recall)
print("F1 Score:", cnn_f1)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_encoder.classes_))

# Save classification report
with open(os.path.join(output_dir, "cnn_classification_report.txt"), "w") as f:
    f.write(classification_report(y_test, y_pred, target_names=target_encoder.classes_))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
plt.imshow(cm)
plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.colorbar()

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xticks([0, 1], target_encoder.classes_)
plt.yticks([0, 1], target_encoder.classes_)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_confusion_matrix.png"))
plt.show()

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, label=f"CNN AUC = {roc_auc:.4f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.title("CNN ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_roc_curve.png"))
plt.show()

# Accuracy and Loss Graphs
plt.figure(figsize=(7, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("CNN Training vs Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_accuracy.png"))
plt.show()

plt.figure(figsize=(7, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CNN Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_loss.png"))
plt.show()

# ============================================================
# Logistic Regression Comparison
# ============================================================

X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(
    X_processed,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_lr, y_train_lr)

lr_pred = logistic_model.predict(X_test_lr)
lr_prob = logistic_model.predict_proba(X_test_lr)[:, 1]

lr_accuracy = accuracy_score(y_test_lr, lr_pred)
lr_precision = precision_score(y_test_lr, lr_pred)
lr_recall = recall_score(y_test_lr, lr_pred)
lr_f1 = f1_score(y_test_lr, lr_pred)

print("\nLogistic Regression Results")
print("Accuracy:", lr_accuracy)
print("Precision:", lr_precision)
print("Recall:", lr_recall)
print("F1 Score:", lr_f1)

# ============================================================
# Model Comparison CSV
# ============================================================

results_df = pd.DataFrame({
    "Model": ["CNN", "Logistic Regression"],
    "Accuracy": [cnn_accuracy, lr_accuracy],
    "Precision": [cnn_precision, lr_precision],
    "Recall": [cnn_recall, lr_recall],
    "F1_Score": [cnn_f1, lr_f1]
})

results_df.to_csv(os.path.join(output_dir, "model_comparison_results.csv"), index=False)

plt.figure(figsize=(8, 5))
x = np.arange(len(results_df["Model"]))
width = 0.2

plt.bar(x - width, results_df["Accuracy"], width, label="Accuracy")
plt.bar(x, results_df["Precision"], width, label="Precision")
plt.bar(x + width, results_df["Recall"], width, label="Recall")
plt.bar(x + 2 * width, results_df["F1_Score"], width, label="F1 Score")

plt.xticks(x, results_df["Model"])
plt.title("CNN vs Logistic Regression Performance")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "model_comparison_graph.png"))
plt.show()

# ============================================================
# Phase 3: K-Fold Cross Validation for CNN
# ============================================================

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results = []

fold = 1

for train_index, val_index in kfold.split(X_images, y):
    print(f"\nTraining Fold {fold}")

    X_train_fold, X_val_fold = X_images[train_index], X_images[val_index]
    y_train_fold, y_val_fold = y.iloc[train_index], y.iloc[val_index]

    fold_model = build_cnn_model(image_size=image_size)

    fold_model.fit(
        X_train_fold,
        y_train_fold,
        epochs=50,
        batch_size=16,
        validation_data=(X_val_fold, y_val_fold),
        callbacks=[early_stop],
        verbose=0
    )

    val_prob = fold_model.predict(X_val_fold).ravel()
    val_pred = (val_prob >= 0.5).astype(int)

    fold_accuracy = accuracy_score(y_val_fold, val_pred)
    fold_precision = precision_score(y_val_fold, val_pred)
    fold_recall = recall_score(y_val_fold, val_pred)
    fold_f1 = f1_score(y_val_fold, val_pred)

    cv_results.append({
        "Fold": fold,
        "Accuracy": fold_accuracy,
        "Precision": fold_precision,
        "Recall": fold_recall,
        "F1_Score": fold_f1
    })

    print("Fold Accuracy:", fold_accuracy)

    fold += 1

cv_results_df = pd.DataFrame(cv_results)
cv_results_df.to_csv(os.path.join(output_dir, "cnn_cross_validation_results.csv"), index=False)

print("\nCross Validation Results:")
print(cv_results_df)

print("\nAverage CV Accuracy:", cv_results_df["Accuracy"].mean())

plt.figure(figsize=(7, 5))
plt.plot(cv_results_df["Fold"], cv_results_df["Accuracy"], marker="o", label="Accuracy")
plt.plot(cv_results_df["Fold"], cv_results_df["F1_Score"], marker="o", label="F1 Score")
plt.title("CNN 5-Fold Cross Validation Performance")
plt.xlabel("Fold")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_cross_validation_graph.png"))
plt.show()

# ============================================================
# Manual Hyperparameter Tuning for CNN
# ============================================================

param_grid = [
    {
        "filters": 16,
        "kernel_size": 2,
        "learning_rate": 0.001,
        "dense_units": 32,
        "dropout_rate": 0.2
    },
    {
        "filters": 32,
        "kernel_size": 3,
        "learning_rate": 0.001,
        "dense_units": 64,
        "dropout_rate": 0.3
    },
    {
        "filters": 64,
        "kernel_size": 3,
        "learning_rate": 0.0005,
        "dense_units": 128,
        "dropout_rate": 0.4
    }
]

tuning_results = []

for params in param_grid:
    print("\nTesting Parameters:", params)

    temp_model = build_cnn_model(
        image_size=image_size,
        filters=params["filters"],
        kernel_size=params["kernel_size"],
        learning_rate=params["learning_rate"],
        dense_units=params["dense_units"],
        dropout_rate=params["dropout_rate"]
    )

    temp_model.fit(
        X_train,
        y_train,
        validation_split=0.2,
        epochs=50,
        batch_size=16,
        callbacks=[early_stop],
        verbose=0
    )

    temp_prob = temp_model.predict(X_test).ravel()
    temp_pred = (temp_prob >= 0.5).astype(int)

    temp_accuracy = accuracy_score(y_test, temp_pred)
    temp_precision = precision_score(y_test, temp_pred)
    temp_recall = recall_score(y_test, temp_pred)
    temp_f1 = f1_score(y_test, temp_pred)

    result = params.copy()
    result.update({
        "Accuracy": temp_accuracy,
        "Precision": temp_precision,
        "Recall": temp_recall,
        "F1_Score": temp_f1
    })

    tuning_results.append(result)

tuning_results_df = pd.DataFrame(tuning_results)
tuning_results_df.to_csv(os.path.join(output_dir, "cnn_hyperparameter_tuning_results.csv"), index=False)

print("\nHyperparameter Tuning Results:")
print(tuning_results_df)

best_result = tuning_results_df.sort_values(by="F1_Score", ascending=False).iloc[0]

print("\nBest CNN Parameters:")
print(best_result)

# ============================================================
# Prediction Output File
# ============================================================

final_predictions = pd.DataFrame({
    "Actual_Label": target_encoder.inverse_transform(y_test),
    "Predicted_Label": target_encoder.inverse_transform(y_pred),
    "Prediction_Probability": y_pred_prob
})

final_predictions.to_csv(os.path.join(output_dir, "cnn_predictions.csv"), index=False)

# ============================================================
# Final Report Generation
# ============================================================

report = f"""
Food Delivery Time Prediction using CNN

Objective:
The goal of this project is to predict whether a food delivery will be Fast or Delayed
based on customer location, restaurant location, weather, traffic conditions, delivery
person experience, order priority, vehicle type, ratings, cost, tip amount, and time-based features.

Dataset:
Input File:
{input_file}

Dataset Shape:
{df.shape}

Target Creation:
The Delivery_Time column was converted into a binary classification target.
Deliveries less than or equal to the median delivery time were labelled as Fast.
Deliveries greater than the median delivery time were labelled as Delayed.

Median Delivery Time:
{median_delivery_time}

Feature Engineering:
1. Customer latitude and longitude were extracted from Customer_Location.
2. Restaurant latitude and longitude were extracted from Restaurant_Location.
3. Haversine distance was calculated between customer and restaurant.
4. Rush hour feature was created using Order_Time.
5. Order hour was extracted from Order_Time.
6. Categorical features were encoded using One-Hot Encoding.
7. Numerical features were scaled using StandardScaler.
8. Missing numerical values were handled using median imputation.
9. Missing categorical values were handled using most frequent imputation.

CNN Method:
Since the dataset is tabular and does not contain real route/map images, the processed
feature vector was converted into an image-like matrix. This allowed CNN layers to process
the structured delivery features.

CNN Results:
Accuracy: {cnn_accuracy}
Precision: {cnn_precision}
Recall: {cnn_recall}
F1 Score: {cnn_f1}
ROC AUC: {roc_auc}

Logistic Regression Results:
Accuracy: {lr_accuracy}
Precision: {lr_precision}
Recall: {lr_recall}
F1 Score: {lr_f1}

Cross Validation:
Average CV Accuracy: {cv_results_df["Accuracy"].mean()}
Average CV F1 Score: {cv_results_df["F1_Score"].mean()}

Best CNN Hyperparameters:
{best_result.to_dict()}

Generated Files:
1. processed_food_delivery_data.csv
2. cnn_food_delivery_model.h5
3. cnn_classification_report.txt
4. cnn_confusion_matrix.png
5. cnn_roc_curve.png
6. cnn_training_accuracy.png
7. cnn_training_loss.png
8. model_comparison_results.csv
9. model_comparison_graph.png
10. cnn_cross_validation_results.csv
11. cnn_cross_validation_graph.png
12. cnn_hyperparameter_tuning_results.csv
13. cnn_predictions.csv
14. final_report.txt

Conclusion:
The CNN model was successfully implemented for Fast vs Delayed food delivery prediction.
The model was evaluated using accuracy, precision, recall, F1-score, confusion matrix,
ROC curve, cross-validation, and hyperparameter tuning. Logistic Regression was also
used as a baseline comparison model.
"""

with open(os.path.join(output_dir, "final_report.txt"), "w") as f:
    f.write(report)

print("\nAll files generated successfully in:")
print(output_dir)
```
