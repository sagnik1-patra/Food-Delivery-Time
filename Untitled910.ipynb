```python
import os
import ast
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_curve, auc
)
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# Paths
# ============================================================

input_file = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time\Food_Delivery_Time_Prediction.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time"

os.makedirs(output_dir, exist_ok=True)

print("Output Directory:", output_dir)

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(input_file)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
display(df.head())

# ============================================================
# Helper Functions
# ============================================================

def parse_location(location):
    try:
        loc = ast.literal_eval(str(location))
        return pd.Series([float(loc[0]), float(loc[1])])
    except:
        return pd.Series([np.nan, np.nan])


def haversine(lat1, lon1, lat2, lon2):
    R = 6371

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        np.sin(dlat / 2) ** 2
        + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    )

    c = 2 * np.arcsin(np.sqrt(a))
    return R * c


def create_rush_hour_feature(time_value):
    try:
        hour = pd.to_datetime(str(time_value)).hour
        if (8 <= hour <= 10) or (18 <= hour <= 21):
            return 1
        return 0
    except:
        return 0


def tabular_to_image(data):
    n_samples, n_features = data.shape
    image_size = int(np.ceil(np.sqrt(n_features)))
    total_pixels = image_size * image_size

    padded_data = np.zeros((n_samples, total_pixels))
    padded_data[:, :n_features] = data

    image_data = padded_data.reshape(n_samples, image_size, image_size, 1)
    return image_data, image_size


def build_cnn_model(
    image_size,
    filters=32,
    kernel_size=3,
    learning_rate=0.001,
    dense_units=64,
    dropout_rate=0.3
):
    model = Sequential()

    model.add(
        Conv2D(
            filters=filters,
            kernel_size=(kernel_size, kernel_size),
            activation="relu",
            padding="same",
            input_shape=(image_size, image_size, 1)
        )
    )
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(
        Conv2D(
            filters=filters * 2,
            kernel_size=(kernel_size, kernel_size),
            activation="relu",
            padding="same"
        )
    )
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(Flatten())
    model.add(Dense(dense_units, activation="relu"))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation="sigmoid"))

    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model

# ============================================================
# Phase 1: Missing Data Handling
# ============================================================

print("\nMissing Values Before Preprocessing:")
print(df.isnull().sum())

if "Order_ID" in df.columns:
    df.drop(columns=["Order_ID"], inplace=True)

# ============================================================
# Feature Engineering
# ============================================================

df[["Customer_Latitude", "Customer_Longitude"]] = df["Customer_Location"].apply(parse_location)
df[["Restaurant_Latitude", "Restaurant_Longitude"]] = df["Restaurant_Location"].apply(parse_location)

df["Haversine_Distance"] = haversine(
    df["Customer_Latitude"],
    df["Customer_Longitude"],
    df["Restaurant_Latitude"],
    df["Restaurant_Longitude"]
)

df["Rush_Hour"] = df["Order_Time"].apply(create_rush_hour_feature)
df["Order_Hour"] = pd.to_datetime(df["Order_Time"], errors="coerce").dt.hour
df["Order_Hour"] = df["Order_Hour"].fillna(df["Order_Hour"].median())

# Weather impact feature
weather_impact_map = {
    "Sunny": 1,
    "Clear": 1,
    "Cloudy": 2,
    "Windy": 3,
    "Fog": 4,
    "Rainy": 5,
    "Stormy": 6
}

if "Weather_Conditions" in df.columns:
    df["Weather_Impact"] = df["Weather_Conditions"].map(weather_impact_map)
    df["Weather_Impact"] = df["Weather_Impact"].fillna(df["Weather_Impact"].median())

# Target creation
median_delivery_time = df["Delivery_Time"].median()

df["Delivery_Status"] = np.where(
    df["Delivery_Time"] <= median_delivery_time,
    "Fast",
    "Delayed"
)

target_encoder = LabelEncoder()
df["Delivery_Status_Encoded"] = target_encoder.fit_transform(df["Delivery_Status"])

print("\nTarget Classes:")
for label, value in zip(target_encoder.classes_, target_encoder.transform(target_encoder.classes_)):
    print(label, "=", value)

# ============================================================
# Remove Raw Text Columns
# ============================================================

drop_columns = [
    "Customer_Location",
    "Restaurant_Location",
    "Order_Time",
    "Delivery_Status"
]

df_model = df.drop(columns=[col for col in drop_columns if col in df.columns])

X = df_model.drop(columns=["Delivery_Status_Encoded"])
y = df_model["Delivery_Status_Encoded"]

numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()

print("\nNumerical Features:", numeric_cols)
print("Categorical Features:", categorical_cols)

# ============================================================
# Encoding + Normalization
# ============================================================

numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

X_processed = preprocessor.fit_transform(X)

if hasattr(X_processed, "toarray"):
    X_processed = X_processed.toarray()

print("Processed Feature Shape:", X_processed.shape)

# ============================================================
# Image-Based Data Representation for CNN
# ============================================================

X_images, image_size = tabular_to_image(X_processed)

print("CNN Image Shape:", X_images.shape)
print("Each sample converted to:", image_size, "x", image_size, "image")

# ============================================================
# Train Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_images,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(
    X_processed,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# ============================================================
# CNN Architecture Implementation
# ============================================================

cnn_model = build_cnn_model(image_size=image_size)
cnn_model.summary()

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

history = cnn_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

# NOTE: h5 model is NOT saved to keep submission small

# ============================================================
# CNN Evaluation
# ============================================================

y_pred_prob = cnn_model.predict(X_test).ravel()
y_pred = (y_pred_prob >= 0.5).astype(int)

cnn_accuracy = accuracy_score(y_test, y_pred)
cnn_precision = precision_score(y_test, y_pred, zero_division=0)
cnn_recall = recall_score(y_test, y_pred, zero_division=0)
cnn_f1 = f1_score(y_test, y_pred, zero_division=0)

print("\nCNN Evaluation Results")
print("Accuracy:", cnn_accuracy)
print("Precision:", cnn_precision)
print("Recall:", cnn_recall)
print("F1 Score:", cnn_f1)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=target_encoder.classes_, zero_division=0))

with open(os.path.join(output_dir, "cnn_classification_report.txt"), "w") as f:
    f.write(classification_report(y_test, y_pred, target_names=target_encoder.classes_, zero_division=0))

# ============================================================
# Confusion Matrix
# ============================================================

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
plt.imshow(cm)
plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.colorbar()

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xticks([0, 1], target_encoder.classes_)
plt.yticks([0, 1], target_encoder.classes_)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_confusion_matrix.png"))
plt.show()

# ============================================================
# ROC Curve
# ============================================================

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, label=f"CNN AUC = {roc_auc:.4f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.title("CNN ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_roc_curve.png"))
plt.show()

# ============================================================
# Training Graphs
# ============================================================

plt.figure(figsize=(7, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("CNN Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_accuracy.png"))
plt.show()

plt.figure(figsize=(7, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CNN Training and Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_loss.png"))
plt.show()

# ============================================================
# Logistic Regression Comparison
# ============================================================

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_lr, y_train_lr)

lr_pred = logistic_model.predict(X_test_lr)
lr_prob = logistic_model.predict_proba(X_test_lr)[:, 1]

lr_accuracy = accuracy_score(y_test_lr, lr_pred)
lr_precision = precision_score(y_test_lr, lr_pred, zero_division=0)
lr_recall = recall_score(y_test_lr, lr_pred, zero_division=0)
lr_f1 = f1_score(y_test_lr, lr_pred, zero_division=0)

print("\nLogistic Regression Results")
print("Accuracy:", lr_accuracy)
print("Precision:", lr_precision)
print("Recall:", lr_recall)
print("F1 Score:", lr_f1)

model_comparison = pd.DataFrame({
    "Model": ["CNN", "Logistic Regression"],
    "Accuracy": [cnn_accuracy, lr_accuracy],
    "Precision": [cnn_precision, lr_precision],
    "Recall": [cnn_recall, lr_recall],
    "F1_Score": [cnn_f1, lr_f1]
})

display(model_comparison)

plt.figure(figsize=(8, 5))
x = np.arange(len(model_comparison["Model"]))
width = 0.2

plt.bar(x - width, model_comparison["Accuracy"], width, label="Accuracy")
plt.bar(x, model_comparison["Precision"], width, label="Precision")
plt.bar(x + width, model_comparison["Recall"], width, label="Recall")
plt.bar(x + 2 * width, model_comparison["F1_Score"], width, label="F1 Score")

plt.xticks(x, model_comparison["Model"])
plt.title("CNN vs Logistic Regression Performance")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "model_comparison_graph.png"))
plt.show()

# ============================================================
# K-Fold Cross Validation
# ============================================================

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results = []
fold = 1

for train_index, val_index in kfold.split(X_images, y):

    print("\nTraining Fold:", fold)

    X_train_fold = X_images[train_index]
    X_val_fold = X_images[val_index]

    y_train_fold = y.iloc[train_index]
    y_val_fold = y.iloc[val_index]

    fold_model = build_cnn_model(image_size=image_size)

    fold_early_stop = EarlyStopping(
        monitor="val_loss",
        patience=8,
        restore_best_weights=True
    )

    fold_model.fit(
        X_train_fold,
        y_train_fold,
        validation_data=(X_val_fold, y_val_fold),
        epochs=50,
        batch_size=16,
        callbacks=[fold_early_stop],
        verbose=0
    )

    val_prob = fold_model.predict(X_val_fold).ravel()
    val_pred = (val_prob >= 0.5).astype(int)

    cv_results.append({
        "Fold": fold,
        "Accuracy": accuracy_score(y_val_fold, val_pred),
        "Precision": precision_score(y_val_fold, val_pred, zero_division=0),
        "Recall": recall_score(y_val_fold, val_pred, zero_division=0),
        "F1_Score": f1_score(y_val_fold, val_pred, zero_division=0)
    })

    fold += 1

cv_results_df = pd.DataFrame(cv_results)
display(cv_results_df)

plt.figure(figsize=(7, 5))
plt.plot(cv_results_df["Fold"], cv_results_df["Accuracy"], marker="o", label="Accuracy")
plt.plot(cv_results_df["Fold"], cv_results_df["F1_Score"], marker="o", label="F1 Score")
plt.title("CNN 5-Fold Cross Validation Performance")
plt.xlabel("Fold")
plt.ylabel("Score")
plt.xticks(cv_results_df["Fold"])
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_cross_validation_graph.png"))
plt.show()

# ============================================================
# Hyperparameter Tuning using RandomizedSearchCV
# ============================================================

class KerasCNNClassifier(BaseEstimator, ClassifierMixin):

    def __init__(
        self,
        image_size=None,
        filters=32,
        kernel_size=3,
        learning_rate=0.001,
        dense_units=64,
        dropout_rate=0.3,
        epochs=20,
        batch_size=16,
        verbose=0
    ):
        self.image_size = image_size
        self.filters = filters
        self.kernel_size = kernel_size
        self.learning_rate = learning_rate
        self.dense_units = dense_units
        self.dropout_rate = dropout_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.verbose = verbose

    def fit(self, X, y):
        self.model_ = build_cnn_model(
            image_size=self.image_size,
            filters=self.filters,
            kernel_size=self.kernel_size,
            learning_rate=self.learning_rate,
            dense_units=self.dense_units,
            dropout_rate=self.dropout_rate
        )

        early_stop = EarlyStopping(
            monitor="loss",
            patience=5,
            restore_best_weights=True
        )

        self.model_.fit(
            X,
            y,
            epochs=self.epochs,
            batch_size=self.batch_size,
            verbose=self.verbose,
            callbacks=[early_stop]
        )

        return self

    def predict(self, X):
        prob = self.model_.predict(X, verbose=0).ravel()
        return (prob >= 0.5).astype(int)

    def predict_proba(self, X):
        prob = self.model_.predict(X, verbose=0).ravel()
        return np.vstack([1 - prob, prob]).T


param_distributions = {
    "filters": [16, 32, 64],
    "kernel_size": [2, 3],
    "learning_rate": [0.001, 0.0005],
    "dense_units": [32, 64, 128],
    "dropout_rate": [0.2, 0.3, 0.4],
    "epochs": [20, 30],
    "batch_size": [8, 16]
}

cnn_estimator = KerasCNNClassifier(
    image_size=image_size,
    verbose=0
)

random_search = RandomizedSearchCV(
    estimator=cnn_estimator,
    param_distributions=param_distributions,
    n_iter=3,
    cv=3,
    scoring="f1",
    random_state=42,
    verbose=2,
    n_jobs=1
)

random_search.fit(X_images, y)

print("\nBest Parameters from RandomizedSearchCV:")
print(random_search.best_params_)

print("\nBest F1 Score from RandomizedSearchCV:")
print(random_search.best_score_)

tuning_results_df = pd.DataFrame(random_search.cv_results_)
display(tuning_results_df[["params", "mean_test_score", "rank_test_score"]])

# ============================================================
# Final Report
# ============================================================

final_report = f"""
Food Delivery Time Prediction using CNN and Model Validation

Author:
Sagnik Patra

Objective:
Predict whether a food delivery is Fast or Delayed using location, weather, traffic,
vehicle, rating, order priority, distance, and time-based features.

Dataset Loaded:
Food_Delivery_Time_Prediction.csv

Dataset Shape:
{df.shape}

Preprocessing Completed:
1. Dataset loaded successfully.
2. Missing numerical values handled using median imputation.
3. Missing categorical values handled using most frequent imputation.
4. Categorical features encoded using One-Hot Encoding.
5. Numerical features normalized using StandardScaler.

Feature Engineering Completed:
1. Customer coordinates extracted.
2. Restaurant coordinates extracted.
3. Haversine distance calculated.
4. Rush hour feature created.
5. Order hour feature created.
6. Weather impact feature created.

Image-Based CNN Representation:
Tabular features were converted into square image-like matrices so that Conv2D CNN
layers could process the data.

CNN Model:
Conv2D, BatchNormalization, MaxPooling2D, Flatten, Dense, Dropout, and Sigmoid layers.

CNN Evaluation:
Accuracy: {cnn_accuracy}
Precision: {cnn_precision}
Recall: {cnn_recall}
F1 Score: {cnn_f1}
ROC AUC: {roc_auc}

Logistic Regression Comparison:
Accuracy: {lr_accuracy}
Precision: {lr_precision}
Recall: {lr_recall}
F1 Score: {lr_f1}

Cross Validation:
Average CV Accuracy: {cv_results_df["Accuracy"].mean()}
Average CV Precision: {cv_results_df["Precision"].mean()}
Average CV Recall: {cv_results_df["Recall"].mean()}
Average CV F1 Score: {cv_results_df["F1_Score"].mean()}

Hyperparameter Tuning:
Method: RandomizedSearchCV
Best Parameters: {random_search.best_params_}
Best F1 Score: {random_search.best_score_}

Visualizations Generated:
1. cnn_confusion_matrix.png
2. cnn_roc_curve.png
3. cnn_training_accuracy.png
4. cnn_training_loss.png
5. cnn_cross_validation_graph.png
6. model_comparison_graph.png

Conclusion:
All required assignment components were completed, including preprocessing, missing
value handling, categorical encoding, normalization, Haversine feature engineering,
rush-hour and weather features, image-based CNN representation, CNN implementation,
training, evaluation, K-fold cross-validation, ROC curve, confusion matrix,
RandomizedSearchCV tuning, and Logistic Regression comparison.
"""

with open(os.path.join(output_dir, "final_report.txt"), "w") as f:
    f.write(final_report)

print(final_report)

# ============================================================
# Submission Instruction
# ============================================================

print("\nSUBMIT THESE FILES ONLY:")
print("1. Food_Delivery_Time_CNN.ipynb  ← Must contain this full executed notebook")
print("2. README.md")
print("3. final_report.txt")
print("4. cnn_confusion_matrix.png")
print("5. cnn_roc_curve.png")
print("6. cnn_training_accuracy.png")
print("7. cnn_training_loss.png")
print("8. cnn_cross_validation_graph.png")
print("9. model_comparison_graph.png")

print("\nDO NOT SUBMIT:")
print("Food_Delivery_Time_Prediction.csv")
print("cnn_food_delivery_model.h5")
print("processed CSV files")
print("venv / env / node_modules")
print("__pycache__ / .cache / .ipynb_checkpoints")
```
