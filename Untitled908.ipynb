```python
import os
import ast
import json
import shutil
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_curve, auc
)
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# Paths
# ============================================================

input_file = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time\Food_Delivery_Time_Prediction.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\Food Delivery Time"

os.makedirs(output_dir, exist_ok=True)

# ============================================================
# Remove Unwanted Submission Files/Folders
# ============================================================

excluded_files = [
    "processed_food_delivery_data.csv",
    "dataset_preview.csv"
]

excluded_folders = [
    "venv",
    "env",
    ".venv",
    "node_modules",
    "dist",
    "build",
    ".next",
    ".cache",
    "__pycache__",
    ".ipynb_checkpoints"
]

for file in excluded_files:
    path = os.path.join(output_dir, file)
    if os.path.exists(path):
        os.remove(path)

for folder in excluded_folders:
    path = os.path.join(output_dir, folder)
    if os.path.exists(path):
        shutil.rmtree(path)

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(input_file)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# ============================================================
# Helper Functions
# ============================================================

def parse_location(location):
    try:
        loc = ast.literal_eval(str(location))
        return pd.Series([float(loc[0]), float(loc[1])])
    except:
        return pd.Series([np.nan, np.nan])


def haversine(lat1, lon1, lat2, lon2):
    R = 6371

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        np.sin(dlat / 2) ** 2
        + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    )

    c = 2 * np.arcsin(np.sqrt(a))
    return R * c


def create_rush_hour_feature(time_value):
    try:
        hour = pd.to_datetime(str(time_value)).hour
        if (8 <= hour <= 10) or (18 <= hour <= 21):
            return 1
        return 0
    except:
        return 0


def tabular_to_image(data):
    n_samples, n_features = data.shape
    image_size = int(np.ceil(np.sqrt(n_features)))
    total_pixels = image_size * image_size

    padded_data = np.zeros((n_samples, total_pixels))
    padded_data[:, :n_features] = data

    image_data = padded_data.reshape(n_samples, image_size, image_size, 1)
    return image_data, image_size


def build_cnn_model(
    image_size,
    filters=32,
    kernel_size=3,
    learning_rate=0.001,
    dense_units=64,
    dropout_rate=0.3
):
    model = Sequential()

    model.add(
        Conv2D(
            filters=filters,
            kernel_size=(kernel_size, kernel_size),
            activation="relu",
            padding="same",
            input_shape=(image_size, image_size, 1)
        )
    )

    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(
        Conv2D(
            filters=filters * 2,
            kernel_size=(kernel_size, kernel_size),
            activation="relu",
            padding="same"
        )
    )

    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2), padding="same"))

    model.add(Flatten())
    model.add(Dense(dense_units, activation="relu"))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation="sigmoid"))

    optimizer = Adam(learning_rate=learning_rate)

    model.compile(
        optimizer=optimizer,
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model


def clear_notebook_outputs(notebook_path):
    if not os.path.exists(notebook_path):
        return

    with open(notebook_path, "r", encoding="utf-8") as f:
        notebook = json.load(f)

    for cell in notebook.get("cells", []):
        if cell.get("cell_type") == "code":
            cell["outputs"] = []
            cell["execution_count"] = None

    with open(notebook_path, "w", encoding="utf-8") as f:
        json.dump(notebook, f, indent=2)

    print("Notebook outputs cleared:", notebook_path)

# ============================================================
# Phase 1: Data Preprocessing and Feature Engineering
# ============================================================

print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

if "Order_ID" in df.columns:
    df.drop(columns=["Order_ID"], inplace=True)

if "Customer_Location" in df.columns:
    df[["Customer_Latitude", "Customer_Longitude"]] = df["Customer_Location"].apply(parse_location)

if "Restaurant_Location" in df.columns:
    df[["Restaurant_Latitude", "Restaurant_Longitude"]] = df["Restaurant_Location"].apply(parse_location)

if all(col in df.columns for col in [
    "Customer_Latitude",
    "Customer_Longitude",
    "Restaurant_Latitude",
    "Restaurant_Longitude"
]):
    df["Haversine_Distance"] = haversine(
        df["Customer_Latitude"],
        df["Customer_Longitude"],
        df["Restaurant_Latitude"],
        df["Restaurant_Longitude"]
    )

if "Order_Time" in df.columns:
    df["Rush_Hour"] = df["Order_Time"].apply(create_rush_hour_feature)
    df["Order_Hour"] = pd.to_datetime(df["Order_Time"], errors="coerce").dt.hour
    df["Order_Hour"] = df["Order_Hour"].fillna(df["Order_Hour"].median())

median_delivery_time = df["Delivery_Time"].median()

df["Delivery_Status"] = np.where(
    df["Delivery_Time"] <= median_delivery_time,
    "Fast",
    "Delayed"
)

target_encoder = LabelEncoder()
df["Delivery_Status_Encoded"] = target_encoder.fit_transform(df["Delivery_Status"])

drop_columns = [
    "Customer_Location",
    "Restaurant_Location",
    "Order_Time",
    "Delivery_Status"
]

existing_drop_columns = [col for col in drop_columns if col in df.columns]

df_model = df.drop(columns=existing_drop_columns)

X = df_model.drop(columns=["Delivery_Status_Encoded"])
y = df_model["Delivery_Status_Encoded"]

numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()

numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

X_processed = preprocessor.fit_transform(X)

if hasattr(X_processed, "toarray"):
    X_processed = X_processed.toarray()

X_images, image_size = tabular_to_image(X_processed)

# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_images,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(
    X_processed,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# ============================================================
# Phase 2: CNN Model
# ============================================================

cnn_model = build_cnn_model(image_size=image_size)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

history = cnn_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

cnn_model.save(os.path.join(output_dir, "cnn_food_delivery_model.h5"))

# ============================================================
# CNN Evaluation
# ============================================================

y_pred_prob = cnn_model.predict(X_test).ravel()
y_pred = (y_pred_prob >= 0.5).astype(int)

cnn_accuracy = accuracy_score(y_test, y_pred)
cnn_precision = precision_score(y_test, y_pred)
cnn_recall = recall_score(y_test, y_pred)
cnn_f1 = f1_score(y_test, y_pred)

report_text = classification_report(
    y_test,
    y_pred,
    target_names=target_encoder.classes_
)

with open(os.path.join(output_dir, "cnn_classification_report.txt"), "w") as f:
    f.write(report_text)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
plt.imshow(cm)
plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.colorbar()

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xticks([0, 1], target_encoder.classes_)
plt.yticks([0, 1], target_encoder.classes_)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_confusion_matrix.png"))
plt.show()

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, label=f"CNN AUC = {roc_auc:.4f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.title("CNN ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_roc_curve.png"))
plt.show()

plt.figure(figsize=(7, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("CNN Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_accuracy.png"))
plt.show()

plt.figure(figsize=(7, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CNN Training and Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_training_loss.png"))
plt.show()

# ============================================================
# Logistic Regression Baseline
# ============================================================

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_lr, y_train_lr)

lr_pred = logistic_model.predict(X_test_lr)

lr_accuracy = accuracy_score(y_test_lr, lr_pred)
lr_precision = precision_score(y_test_lr, lr_pred)
lr_recall = recall_score(y_test_lr, lr_pred)
lr_f1 = f1_score(y_test_lr, lr_pred)

model_comparison = pd.DataFrame(
    {
        "Model": ["CNN", "Logistic Regression"],
        "Accuracy": [cnn_accuracy, lr_accuracy],
        "Precision": [cnn_precision, lr_precision],
        "Recall": [cnn_recall, lr_recall],
        "F1_Score": [cnn_f1, lr_f1]
    }
)

model_comparison.to_csv(
    os.path.join(output_dir, "model_comparison_results.csv"),
    index=False
)

plt.figure(figsize=(8, 5))

x = np.arange(len(model_comparison["Model"]))
width = 0.2

plt.bar(x - width, model_comparison["Accuracy"], width, label="Accuracy")
plt.bar(x, model_comparison["Precision"], width, label="Precision")
plt.bar(x + width, model_comparison["Recall"], width, label="Recall")
plt.bar(x + 2 * width, model_comparison["F1_Score"], width, label="F1 Score")

plt.xticks(x, model_comparison["Model"])
plt.title("CNN vs Logistic Regression Performance")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "model_comparison_graph.png"))
plt.show()

# ============================================================
# Phase 3: 5-Fold CNN Cross Validation
# ============================================================

kfold = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

cv_results = []
fold = 1

for train_index, val_index in kfold.split(X_images, y):

    X_train_fold = X_images[train_index]
    X_val_fold = X_images[val_index]

    y_train_fold = y.iloc[train_index]
    y_val_fold = y.iloc[val_index]

    fold_model = build_cnn_model(image_size=image_size)

    fold_early_stop = EarlyStopping(
        monitor="val_loss",
        patience=8,
        restore_best_weights=True
    )

    fold_model.fit(
        X_train_fold,
        y_train_fold,
        validation_data=(X_val_fold, y_val_fold),
        epochs=50,
        batch_size=16,
        callbacks=[fold_early_stop],
        verbose=0
    )

    val_prob = fold_model.predict(X_val_fold).ravel()
    val_pred = (val_prob >= 0.5).astype(int)

    cv_results.append(
        {
            "Fold": fold,
            "Accuracy": accuracy_score(y_val_fold, val_pred),
            "Precision": precision_score(y_val_fold, val_pred),
            "Recall": recall_score(y_val_fold, val_pred),
            "F1_Score": f1_score(y_val_fold, val_pred)
        }
    )

    fold += 1

cv_results_df = pd.DataFrame(cv_results)

cv_results_df.to_csv(
    os.path.join(output_dir, "cnn_cross_validation_results.csv"),
    index=False
)

plt.figure(figsize=(7, 5))
plt.plot(cv_results_df["Fold"], cv_results_df["Accuracy"], marker="o", label="Accuracy")
plt.plot(cv_results_df["Fold"], cv_results_df["F1_Score"], marker="o", label="F1 Score")
plt.title("CNN 5-Fold Cross Validation Performance")
plt.xlabel("Fold")
plt.ylabel("Score")
plt.xticks(cv_results_df["Fold"])
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "cnn_cross_validation_graph.png"))
plt.show()

# ============================================================
# CNN Hyperparameter Tuning
# ============================================================

param_grid = [
    {
        "filters": 16,
        "kernel_size": 2,
        "learning_rate": 0.001,
        "dense_units": 32,
        "dropout_rate": 0.2
    },
    {
        "filters": 32,
        "kernel_size": 3,
        "learning_rate": 0.001,
        "dense_units": 64,
        "dropout_rate": 0.3
    },
    {
        "filters": 64,
        "kernel_size": 3,
        "learning_rate": 0.0005,
        "dense_units": 128,
        "dropout_rate": 0.4
    }
]

tuning_results = []

for params in param_grid:

    temp_model = build_cnn_model(
        image_size=image_size,
        filters=params["filters"],
        kernel_size=params["kernel_size"],
        learning_rate=params["learning_rate"],
        dense_units=params["dense_units"],
        dropout_rate=params["dropout_rate"]
    )

    temp_early_stop = EarlyStopping(
        monitor="val_loss",
        patience=8,
        restore_best_weights=True
    )

    temp_model.fit(
        X_train,
        y_train,
        validation_split=0.2,
        epochs=50,
        batch_size=16,
        callbacks=[temp_early_stop],
        verbose=0
    )

    temp_prob = temp_model.predict(X_test).ravel()
    temp_pred = (temp_prob >= 0.5).astype(int)

    result = params.copy()
    result.update(
        {
            "Accuracy": accuracy_score(y_test, temp_pred),
            "Precision": precision_score(y_test, temp_pred),
            "Recall": recall_score(y_test, temp_pred),
            "F1_Score": f1_score(y_test, temp_pred)
        }
    )

    tuning_results.append(result)

tuning_results_df = pd.DataFrame(tuning_results)

tuning_results_df.to_csv(
    os.path.join(output_dir, "cnn_hyperparameter_tuning_results.csv"),
    index=False
)

best_result = tuning_results_df.sort_values(
    by="F1_Score",
    ascending=False
).iloc[0]

# ============================================================
# Predictions
# ============================================================

final_predictions = pd.DataFrame(
    {
        "Actual_Label": target_encoder.inverse_transform(y_test),
        "Predicted_Label": target_encoder.inverse_transform(y_pred),
        "Prediction_Probability": y_pred_prob
    }
)

final_predictions.to_csv(
    os.path.join(output_dir, "cnn_predictions.csv"),
    index=False
)

# ============================================================
# Final Report
# ============================================================

final_report = f"""
Food Delivery Time Prediction using CNN and Model Validation

Author:
Sagnik Patra

Objective:
The goal of this project is to predict whether a food delivery will be Fast or Delayed
using delivery, location, weather, traffic, rating, vehicle, priority, and time-based features.

Dataset:
Food_Delivery_Time_Prediction.csv

Note:
The dataset is used only for model training and is not included in the final submission package.

Dataset Shape:
{df.shape}

Median Delivery Time:
{median_delivery_time}

Target Creation:
Fast = Delivery_Time <= Median Delivery_Time
Delayed = Delivery_Time > Median Delivery_Time

Preprocessing:
1. Missing numerical values were handled using median imputation.
2. Missing categorical values were handled using most frequent imputation.
3. Categorical variables were encoded using One-Hot Encoding.
4. Numerical variables were normalized using StandardScaler.

Feature Engineering:
1. Customer and restaurant coordinates were extracted.
2. Haversine distance was calculated.
3. Rush hour feature was created.
4. Order hour was extracted.

CNN Results:
Accuracy: {cnn_accuracy}
Precision: {cnn_precision}
Recall: {cnn_recall}
F1 Score: {cnn_f1}
ROC AUC: {roc_auc}

Logistic Regression Results:
Accuracy: {lr_accuracy}
Precision: {lr_precision}
Recall: {lr_recall}
F1 Score: {lr_f1}

Cross Validation:
Average Accuracy: {cv_results_df["Accuracy"].mean()}
Average Precision: {cv_results_df["Precision"].mean()}
Average Recall: {cv_results_df["Recall"].mean()}
Average F1 Score: {cv_results_df["F1_Score"].mean()}

Best CNN Hyperparameters:
{best_result.to_dict()}

Required Submission Files:
1. Food_Delivery_Time_CNN.ipynb
2. README.md
3. final_report.txt
4. cnn_food_delivery_model.h5
5. cnn_classification_report.txt
6. model_comparison_results.csv
7. cnn_cross_validation_results.csv
8. cnn_hyperparameter_tuning_results.csv
9. cnn_predictions.csv
10. cnn_confusion_matrix.png
11. cnn_roc_curve.png
12. cnn_training_accuracy.png
13. cnn_training_loss.png
14. cnn_cross_validation_graph.png
15. model_comparison_graph.png

Excluded From Submission:
1. Food_Delivery_Time_Prediction.csv
2. Raw data dumps
3. Large media files
4. venv, env, node_modules
5. dist, build
6. .next, .cache, __pycache__
7. Notebook large outputs

Conclusion:
The CNN model was successfully trained, evaluated, cross-validated, and compared with
Logistic Regression. Only assignment-required outputs are generated and retained.
"""

with open(os.path.join(output_dir, "final_report.txt"), "w") as f:
    f.write(final_report)

# ============================================================
# Optional: Clear Notebook Outputs Before Submission
# ============================================================

notebook_path = os.path.join(output_dir, "Food_Delivery_Time_CNN.ipynb")
clear_notebook_outputs(notebook_path)

# ============================================================
# Final Submission Checklist
# ============================================================

required_submission_files = [
    "Food_Delivery_Time_CNN.ipynb",
    "README.md",
    "final_report.txt",
    "cnn_food_delivery_model.h5",
    "cnn_classification_report.txt",
    "model_comparison_results.csv",
    "cnn_cross_validation_results.csv",
    "cnn_hyperparameter_tuning_results.csv",
    "cnn_predictions.csv",
    "cnn_confusion_matrix.png",
    "cnn_roc_curve.png",
    "cnn_training_accuracy.png",
    "cnn_training_loss.png",
    "cnn_cross_validation_graph.png",
    "model_comparison_graph.png"
]

print("\nPROJECT COMPLETED SUCCESSFULLY")
print("Submit only these files:")

for file in required_submission_files:
    file_path = os.path.join(output_dir, file)

    if os.path.exists(file_path):
        print("Available:", file)
    else:
        print("Missing/Create manually:", file)

print("\nDo NOT submit:")
print("Food_Delivery_Time_Prediction.csv")
print("processed_food_delivery_data.csv")
print("dataset_preview.csv")
print("venv, env, node_modules")
print("dist, build")
print(".next, .cache, __pycache__")
print("large notebook outputs")
```
